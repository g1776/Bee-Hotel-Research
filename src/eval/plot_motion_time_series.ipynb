```python
import plotly.express as px
from datetime import datetime



def plot_motion_frame_histogram(log: str, n_bins: int):
    """Plot motion time series from log file

    Args:
        log (str): Path to log file
    """

    with open(log, "r") as f:
        lines = f.readlines()

    timestamps = []
    for line in lines:
        if line.startswith("---"):
            continue
        timestamp = datetime.strptime(line.split(" ")[-1].strip(), "%H:%M:%S")
        timestamps.append(timestamp)
    
    
    fig = px.histogram(x=timestamps, nbins=n_bins, title=f"\"{log}\" Motion Events Histogram")
    
    # add x label
    fig.update_layout(
        xaxis_title="Timestamp",
        yaxis_title="Number of motion events",
        xaxis_tickformat="%H:%M:%S"
    )

    fig.show()
```

```python
plot_motion_frame_histogram("10am.log", n_bins=100)
```

```python
plot_motion_frame_histogram("11am.log", n_bins=100)
```

```python
plot_motion_frame_histogram("12pm.log", n_bins=100)
```
